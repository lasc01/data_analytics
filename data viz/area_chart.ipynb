```python
# An area chart is used to display quantitative data for 3 or more variable.
# We have the classic area chart, the stack area chart, 100% stacked chart and the steam chart
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set()
```

```python
car_fuel_eng = pd.read_csv('stacked_area_chart_data.csv')
```

```python
car_fuel_eng
```

```python
colors = ["#011638", "#7e2987", "#ef2026"]
label = ["Gas", "Petrol", "Diesel"]
sns.set_style("white")
# This will change the background color to white
plt.figure(figsize = (12, 6))
# This increases the size of the graph
plt.stackplot(car_fuel_eng["Year"],
             car_fuel_eng["Gas"],
             car_fuel_eng["Petrol"],
             car_fuel_eng["Diesel"],
             colors = colors,
             edgecolor = "none")
plt.xticks(car_fuel_eng["Year"], rotation = 60)
plt.legend(labels = label, loc = "upper left")
plt.ylabel("Fuel type engine", fontsize = 13)
plt.title("Popularity of Fuel Type Engine (1982 to 2016)", fontsize = 14, weight = "bold")
sns.despine()
# this will remove the up and right border line. It can also be specified by using right = True, up = True and so on
plt.show()
```
