```python
# Scatter plot shows the relationship between 2 numerical feature

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
real_estate_df = pd.read_csv("scatter_data.csv")
```

```python
real_estate_df
```

```python
plt.figure(figsize = (12,8))
scatter = plt.scatter(real_estate_df["Area (ft.)"],
           real_estate_df["Price"],
           alpha = 0.6,
            # the alpha is used to denote transparency. The transparency ranges from 0 to 1
            c = real_estate_df["Building Type"],
            # The c gives a default color to the building column
            cmap = "mako"
            # The cmap helps to specify the collect of colours we want. We have others like viridis and so on
           )
plt.legend(*scatter.legend_elements(),
          loc = "upper left",
          title = "Building Type")
plt.title("Relationship netween Areas and Price in a Real Estate", fontsize = 14, weight = "bold")
plt.xlabel("Area (ft.)", weight = "bold")
plt.ylabel("Price (000' $)", weight = "bold")
plt.show()
```

```python
# We can also use the power of seaborn to achieve more accurate scatterd chart
plt.figure(figsize = (12,8))
sns.scatterplot(x = real_estate_df["Price"],
               y = real_estate_df["Area (ft.)"],
               hue = real_estate_df["Building Type"],
               # the hue serves as to give colour to the scatter plot using the building type column(just like the c in matplotlin)
               palette = ["blue", "green", "yellow", "red", "brown"],
                # the palette helps to specify colour like the cmap
                s = 100
                # The s parameter helps to increase the size of point
               )
plt.title("Relationship netween Areas and Price in a Real Estate", fontsize = 14, weight = "bold")
plt.xlabel("Area (ft.)", weight = "bold")
plt.ylabel("Price (000' $)", weight = "bold")
plt.show()
```
