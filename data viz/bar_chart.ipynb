```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_cars = pd.read_csv("bar_chart_data.csv")
```

```python
df_used_cars
```

```python
plt.figure(figsize = (9,6))
# note that the default size is 6.4 by 4.8 inches
# We can also achieve the same solution by rotating them using the xticks below
plt.bar(x=df_used_cars['Brand'], 
        height=df_used_cars['Cars Listings'],
       color = "midnightblue")
plt.xticks(rotation = 45, fontsize = 16)
plt.yticks(fontsize = 16)
# We can also improve the appearance of the chat using the above methods. This is the best way of utilizing space
plt.title("Car Brands Chart", fontsize = 16, fontweight = "bold")
plt.ylabel("Car Listing", fontsize = 16, fontweight = "bold")
plt.show
# To save file as png
plt.savefig("Used Car Bars.png")
```
