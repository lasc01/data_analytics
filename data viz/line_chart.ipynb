```python
# Line charts are usually used to represent time series data
# It is also used to track the progression of a chart(3 - 10) over a particular set of time
# It can also be used to track several different variables on the same chart
# It is not necessary to start yoyr y axis at 0 unlike most charts
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_stock_market_indices = pd.read_csv("line_chart_data.csv")
```

```python
df_stock_market_indices
```

```python
# The date is in a string format. Therefore, we have to convert it to date time format using pandas
df_stock_market_indices["new_date"] = pd.to_datetime(df_stock_market_indices["Date"])
```

```python
df_stock_market_indices["new_date"]
```

```python
label = ["GSPC500", "FTSE100"]
plt.figure(figsize = (20, 8))
plt.plot(df_stock_market_indices["new_date"], df_stock_market_indices["GSPC500"])
plt.plot(df_stock_market_indices["new_date"], df_stock_market_indices["FTSE100"])
plt.legend(labels = label, loc = "upper left", fontsize = "larger")
plt.title("FTSE100 & GSPC500 returns(2000-2010)", fontsize = 14, weight = "bold")
plt.ylabel("Returns", fontsize = 14)
plt.xlabel("Date", fontsize = 14)          
plt.show()
```

```python
# The points in the chart are too much to be visible. 
# Let us limit our data frame to the second half of the year 2008
df_stock_market_indices_08_H2 = df_stock_market_indices[(df_stock_market_indices.new_date >= "2008-07-01") & 
                                (df_stock_market_indices.new_date <= "2008-12-31")]
```

```python
df_stock_market_indices_08_H2
```

```python
label = ["GSPC500", "FTSE100"]
plt.figure(figsize = (20, 8))
plt.plot(df_stock_market_indices_08_H2["new_date"], df_stock_market_indices_08_H2["GSPC500"], color = "midnightblue")
plt.plot(df_stock_market_indices_08_H2["new_date"], df_stock_market_indices_08_H2["FTSE100"], color = "darkred")
plt.legend(labels = label, loc = "upper left", fontsize = "larger")
plt.title("FTSE100 & GSPC500 returns(2008 H2)", fontsize = 14, weight = "bold")
plt.ylabel("Returns", fontsize = 14)
plt.xlabel("Date", fontsize = 14)          
plt.show()
```
