```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
# Avoid using piechart
```

```python
df = pd.read_csv("pie_chart_data.csv")
```

```python
df
```

```python
sns.set_palette('colorblind')
```

```python
plt.figure(figsize = (10,8))
plt.pie(df['Number of Cars'],
        labels = df['Engine Fuel Type'].values,
        autopct = '%.2f%%',
        textprops = {"size" : "x-large",
                    "fontweight": "bold",
                    "rotation": 30,
                    "color" : "white"})
plt.legend()
plt.title("Cars by Engine Fuel Type", fontsize = 18, fontweight = "bold")
plt.show()
```
