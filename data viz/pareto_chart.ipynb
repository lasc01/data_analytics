```python
# Pareto chart is a combination of bar and line chart
# The pareto usually have 2 y axis. One is the bar chart axis while the other takes the line chart which is also the cummulative frequency
# The line chart is always in %

import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
```

```python
df_user_surver = pd.read_csv("bar_line_chart_data.csv")
```

```python
df_user_surver
```

```python
fig, ax = plt.subplots(figsize = (12, 8))

ax.bar(df_user_surver["Year"],
      df_user_surver["Participants"],
      color = "k")
ax.set_ylabel("Number of Participants", weight = "bold")
ax.set_xlabel("Years", weight = "bold")
ax.tick_params(axis = "y", width = 2, labelsize = "large")
ax1 = ax.twinx()
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.set_ylabel("Python Users", weight = "bold")
ax1.tick_params(axis = "y", width = 2, labelsize = "large")
ax1.plot(df_user_surver["Year"],
        df_user_surver["Python Users"],
        color = "darkred",
        marker = "D")
ax.set_title("KD Nuggets survey of python users between 2012 - 2019", weight = "bold")
```
