```python
# An histogram shows the distribution of a numeric variable
#  (x...y) This means x and y are excluded. E.g (1, 5) means 2,3,4
# [x...y] This means x and y are included. E.g [1, 5] means 1,2,3,4,5
# (x...y] This means x is excluded but y is included. E.g (1, 5] means 2,3,4,5

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
real_estate_df = pd.read_csv("histogram_data.csv")
```

```python
real_estate_df
```

```python
sns.set_style("dark")
plt.figure(figsize = (8,6))
plt.hist(real_estate_df["Price"],
        bins = 7,
        color = "darkgreen")
# Choosing the number of bins(intervals)is the most crucial part of creating a histogram
plt.title("REAL ESTATE PRICE DISTRIBUTION", fontsize = 15, weight = "bold")
plt.xlabel("Price in (000' $)", weight = "bold")
plt.ylabel("Number of Properties", weight = "bold")
sns.despine()
plt.show()
```
