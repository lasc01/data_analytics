```python
#Regression analysis is a statistical approach used to estimate the relationship between a dependent variable(outcome) and an independent (predictors)
# Regression line can be linear, exponential, polynomial and logarithmic. But we will focus more on linear 

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_sales_budget = pd.read_csv("scatter_plot_ii.csv")
```

```python
df_sales_budget
```

```python
# plt.figure(figsize = (12,8))
# We can also use seaborn to select our figsize
sns.set(rc= {"figure.figsize" : (10,8)})
sns.regplot(
    # we can also use sns.lmplot
    x = df_sales_budget["Budget"],
    y = df_sales_budget["Sales"],
    # for colors
    scatter_kws = {'color' : "black"},
    line_kws = {"color" : "red"}
)
plt.title("Market Sales and Budget", fontsize = 14, weight = "bold")
plt.ylabel("Budget in (000' $)")
plt.xlabel("Sales in (000' units)")
plt.show()
```
